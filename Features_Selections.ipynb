```python
import pandas as pd
import numpy as np

import os
path = os.path.expanduser("~/Documents/useful_df_car_accident_severity.csv")
df = pd.read_csv (path)

df.head(2)
```

```python
X = df[['X', 'Y',
        'ADDRTYPE', 'INTKEY',
        'COLLISIONTYPE', 'PERSONCOUNT',
        'PEDCOUNT', 'PEDCYLCOUNT', 'VEHCOUNT',
        'JUNCTIONTYPE', 'SDOT_COLCODE', 'INATTENTIONIND', 
        'UNDERINFL', 'WEATHER', 'ROADCOND', 'LIGHTCOND',
        'PEDROWNOTGRNT', 'SPEEDING',
        'SEGLANEKEY', 'CROSSWALKKEY', 'HITPARKEDCAR',
        'year', 'month', 'day', 'hour', 'minute', 'weekday_name']]

y = df['SEVERITYCODE']
```

# DecisionTreeClassifier

```python


X = ['X', 'Y', 'SEGLANEKEY', 'ROADCOND', 'LIGHTCOND', 'hour', 'weekday_name']    # type list !!!

y = df['SEVERITYCODE']    # df


def get_combinations(X):
    '''get all combinations of X features'''
    import itertools
    global combinations
    combinations = []
    for L in range(0, len(X)+1):
         for subset in itertools.combinations(X, L):
            subset = list(subset)
            combinations.append(subset)
    del combinations[0]
    return combinations


def iterative_evaluation(df, y, combinations):
    '''iterative evaluation of all features'''
    comand = input(f'You have {len(combinations)} combinations of X features.'
                   f' Input y/n for continuous (+ press enter/return)')
    if comand == 'n':
        return 'OK'
    elif comand == 'y':
# Imports
        from sklearn.model_selection import train_test_split
        from sklearn.tree import DecisionTreeClassifier
        from sklearn import metrics
# Classifier Entity        
        Tree = DecisionTreeClassifier(criterion="entropy")
        
        global evaluations
        evaluations = []
        i=1
        
        for el in combinations:
            print(f'Inprogress {i}/{len(combinations)}')
            el = list(el)
            X = df[el]            
# Model evaluation
            X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.1, random_state=3)
            Tree.fit(X_trainset,y_trainset)            
            predTree = Tree.predict(X_testset)
            accuracy = metrics.accuracy_score(y_testset, predTree)
            
            evaluations.append([accuracy, el])
            i+=1             
        return get_sorted_evaluations(evaluations)
    else:
        return 'False'

    
def get_sorted_evaluations(evaluations):
    '''sort and show list of evaluations'''
    evaluations.sort(reverse=True)
    print("")
    print("___DecisionTreeClassifier  SORTED  EVALUATIONS:___")
    print("")
    print(f"{len(evaluations)} evaluations")
    print("")
    for el in evaluations:
        print(el)
        print("")

        
get_combinations(X) 
iterative_evaluation(df, y, combinations)

```

# LGBMClassifier

```python


# type list !!!
X = ['X', 'Y', 'INTKEY', 'JUNCTIONTYPE', 'WEATHER', 'ROADCOND', 'LIGHTCOND', 'PEDROWNOTGRNT',
         'SEGLANEKEY', 'CROSSWALKKEY', 'year', 'month', 'day', 'hour', 'minute', 'weekday_name']

# type df
y = df['SEVERITYCODE']


def get_combinations(X):
    '''get all combinations of X features'''
    import itertools
    global combinations
    combinations = []
    for L in range(0, len(X)+1):
         for subset in itertools.combinations(X, L):
            subset = list(subset)
            combinations.append(subset)
    del combinations[0]
    return combinations


def iterative_evaluation(df, y, combinations):
    '''iterative evaluation of all features'''
    comand = input(f'You have {len(combinations)} combinations of X features.'
                   f' Input y/n for continuous (+ press enter/return)')
    if comand == 'n':
        return 'OK'
    elif comand == 'y':
# Imports
        import lightgbm as lgb
        from sklearn.model_selection import train_test_split
        from sklearn import metrics
# Classifier Entity        
        lgbm = lgb.LGBMClassifier()
        
        global evaluations
        evaluations = []
        i=1
        
        for el in combinations:
            print(f'Inprogress {i}/{len(combinations)}')
            el = list(el)
            X = df[el]            
# Model evaluation
            X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.1, random_state=3)
    
            lgbm.fit(X_trainset,y_trainset)
            predTree = lgbm.predict(X_testset)
            accuracy = metrics.accuracy_score(y_testset, predTree)
            
            evaluations.append([accuracy, el])
            i+=1             
        return get_sorted_evaluations(evaluations)
    else:
        return 'False'

    
def get_sorted_evaluations(evaluations):
    '''sort and show list of evaluations'''
    evaluations.sort(reverse=True)
    print("")
    print("___DecisionTreeClassifier  SORTED  EVALUATIONS:___")
    print("")
    print(f"{len(evaluations)} evaluations")
    print("")
    for el in evaluations:
        print(el)
        print("")

        
get_combinations(X) 
iterative_evaluation(df, y, combinations)


```
